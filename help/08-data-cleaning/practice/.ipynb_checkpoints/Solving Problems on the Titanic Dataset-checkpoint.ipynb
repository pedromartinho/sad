# Apply to Titanic Dataset

```python
import pandas as pd
import numpy as np

% matplotlib inline
from matplotlib import pyplot as plt 
```

# Load data

```python
df = pd.read_csv('../data/titanic_exercise.csv')
df.head()
df.describe()
```

```python
df.isnull().sum()
```

# Task: Clean outliers, duplicates and missing values

```python
df = df.drop_duplicates(keep='first')
df.head(n=10)
```

```python
df = df.drop('Cabin', axis=1)
```

```python
df['Age'] = df['Age'].replace(np.nan, df['Age'].median())
```

```python
age_done = df[(df['Age'] >= 1) & (df['Age'] < 100)]
age_done.head(n=5)
```

```python
age_done['Embarked'] = age_done['Embarked'].replace(np.nan, age_done['Embarked'].mode()[0])
age_done.head(n=10)
```
