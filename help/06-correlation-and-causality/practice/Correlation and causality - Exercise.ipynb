```python
# import Pandas 
import pandas as pd 
# making plots inline 
% matplotlib inline
```

```python
data = pd.read_csv('../data/stocks_and_airliners.csv', index_col='date')
```

```python
data.shape
```

```python
data.columns
data.shape
```

```python
data['Accenture plc'].corr(data['Adobe Systems Inc'], method='pearson')
data.shape
```

```python
data = data.reset_index(drop=True)
corrs = data.corr(method='spearman')
corrs
```

```python
corrs.stack().head(10)
```

```python
without_the_ones = corrs.stack().loc[corrs.stack() < 1]
without_the_ones.head(10)
```

```python
max_pair = without_the_ones.idxmax()
max_pair
```

```python
corrs.stack().loc[max_pair]
```
