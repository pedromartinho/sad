## Help Diana

```python
worked_diana = data
worked_diana['room_type'] = worked_diana['room_type'].map({'Shared room':'0','Private room':'1','Entire home/apt':'2'})
worked_diana = worked_diana.sort_values(by=['price','room_type','overall_satisfaction','reviews'])
worked_diana['room_type'] = worked_diana['room_type'].map({'0':'Shared room','1':'Private room','2':'Entire home/apt'})
worked_diana.head(n=10)
```

```python
worked_diana.to_json('help_Diana.json',orient='index')
```
