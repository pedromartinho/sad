## Import packages
Pandas to work the data and matplotlib.pyplot for data visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.cluster as cluster
```

```python
%matplotlib notebook
plt.style.use('ggplot')
```

```python
def apply_kmeans(features,n):
    est = cluster.KMeans(n_clusters=n)
    Y = pd.DataFrame(est.fit_predict(features), columns=['cluster ID'])
    Z = pd.DataFrame(est.cluster_centers_[Y['cluster ID']], 
                     columns=features.columns)
    result = pd.concat([features, Y, Z], axis=1) 
    return result
```

### a.

```python
iris = pd.read_csv("data/iris.csv")
iris = iris.drop( ' class ',axis=1)
iris.head(5)
```

```python
 result = apply_kmeans(iris,2)
```
