# NoShows Data

```python
noshows = pd.read_csv("../data/noshows.csv")
print(noshows.shape)
noshows.columns
```

```python
noshows = noshows.drop(['PatientId', 'AppointmentID','ScheduledDay','No-show'],axis=1)
```

```python
noshows['AppointmentDay'] = noshows['AppointmentDay'].apply(lambda x: str(x).split("T")[0])
noshows['AppointmentDay'] = noshows['AppointmentDay'].apply(lambda x: str(x).split("2016-")[1])
noshows.AppointmentDay.unique()
```

```python
print(noshows.info())
noshows.describe()
```

```python
noshows.head(3)
```

```python
noshows.Handcap.plot.hist()
plt.title('Handcap feature')
```

```python
plt.rcParams.update({'font.size': 30})
noshows.AppointmentDay.value_counts()
noshows.AppointmentDay.value_counts().plot(kind='bar')
plt.title('AppointmentDay feature')
```

```python
count_doenca = noshows.Neighbourhood.value_counts()
count_doenca = count_doenca[0:20]
count_doenca.plot(kind='bar')
plt.title('Neighbourhood feature')
```

```python
noshows.Age.plot.hist()
plt.title('Age feature')
```
