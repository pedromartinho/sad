```python
fig, axes = plt.subplots(nrows=3, ncols=2)
ax0, ax1, ax2, ax3, ax4, ax5 = axes.flatten()
ax0.hist(crabs.BD, 8, histtype='bar', stacked=True)
ax0.set_title('BD feature')
ax1.hist(crabs.CL, 8, histtype='bar', stacked=True)
ax1.set_title('CL feature')
ax2.hist(crabs.CW, 8, histtype='bar', stacked=True)
ax2.set_title('CW feature')
ax3.hist(crabs.FL, 8, histtype='bar', stacked=True)
ax3.set_title('FL feature')

ax4.hist(crabs.RW, 8, histtype='bar', stacked=True)
ax4.set_title('RW feature')
ax5.set_visible(False)
fig.subplots_adjust(hspace=0.5)
plt.show()
```
